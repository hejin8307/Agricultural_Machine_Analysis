```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
file = pd.read_csv('C:/Users/user/Desktop/NIA/project1/data/preprocessing/address.csv', parse_dates=["date"], encoding = 'euc-kr')
file = file.drop(columns = 'Unnamed: 0')
file = file.drop(columns = '일시')
file
```

```python
arr = file['주소'].str.split(' ').tolist()
arr
```

```python
add_list = []
for i in range(0, 27299) :
    add_list.append(arr[i][2])
add_list
```

```python
file['읍면동'] =  add_list
file
```

```python
most1 = pd.DataFrame({'count' : file.groupby(['읍면동','구분'])[['읍면동','구분']].size()}).reset_index().sort_values('count',ascending = False).drop_duplicates(subset=['읍면동'], keep='first')
```

```python
most2 = pd.DataFrame({'count' : file.groupby(['읍면동','종류'])[['읍면동','종류']].size()}).reset_index().sort_values('count',ascending = False).drop_duplicates(subset=['읍면동'], keep='first')
```

```python
most1[['구분']].value_counts()
```

```python
most1.to_csv('C:/Users/user/Desktop/NIA/project1/data/preprocessing/가장 많이 사용한 농기계(구분).csv', encoding = 'euc-kr')
```

```python
most2.to_csv('C:/Users/user/Desktop/NIA/project1/data/preprocessing/가장 많이 사용한 농기계(종류).csv', encoding = 'euc-kr')
```
